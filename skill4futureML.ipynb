```python
# Importing necessary libraries
import pandas as pd # data manupulation
import numpy as np # linear algebra & arrays
import matplotlib.pyplot as plt # visualization
```

```python
# Load the dataset
df=pd.read_csv('appliance_energy.csv')
df.head() # to display top 5 rows, to display bottom 5 rows-tail()
```

```python
# To know rows & cols
df.shape
```

```python
# Information
df.info()
```

```python
# Statistical info-dataset
df.describe()
```

```python
# To check the missing values
df.isnull().sum() # if we had missing values we can use drop/impute
```

```python
# Feature and target selection
x=df[['Temperature (°C)']]
y=df['Energy Consumption (kWh)']
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Regression model
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()

# Train the model
lr_model.fit(x_train,y_train)
```

```python
# Prediction
lr_y_pred = lr_model.predict(x_test)
lr_y_pred
```

```python
# Metrics
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,lr_y_pred)
r2=r2_score(y_test,lr_y_pred)
print(f' The Mean Squared Error is: {mse}')
print(f'R-squared: {r2}')
```

```python
# Plot
plt.scatter(x_test,y_test,color='blue',label='Test Data')
plt.plot(x_test,lr_y_pred,color='red',label='Regression Line')
plt.xlabel('Temperature (°C)')
plt.ylabel('Energy Consumption (kWh)')
plt.title('Energy consumption using linear regression')
plt.show()
```

```python
# To save the model
import joblib # Importing joblib library
joblib.dump(lr_model,'energy_consumption_model.pkl')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Load the dataset
df=pd.read_csv('sustainability_data.csv')
df.head()
```

```python
# To see column name
df.columns
```

```python
# Feature selection
x=df[['Year','Energy_Consumption','Renewable_Percentage','GDP']]
y=df['CO2_Emissions']
```

```python
# Split the data
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
# To define the degree of polynomial
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
# Initialize and train the model
poly_model=LinearRegression()
poly_model.fit(x_train_poly,y_train)
```

```python
# Prediction
poly_pred=poly_model.predict(x_test_poly)
poly_pred
```

```python
# Evaluate
mse=mean_squared_error(y_test,poly_pred)
r2=r2_score(y_test,poly_pred)
print(f'The Mean Squared Error is: {mse}')
print(f'The R-squared is: {r2}')
```

```python
import joblib
joblib.dump(poly_model,'sustainability_model.pkl')
```

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df=pd.read_csv('green_tech_data.csv')
df.head(10)# Top 10 rows
```

```python
# Find the missing values
df.isnull().sum()
```

```python
# NO. of rows and columns
df.shape
```

```python
# Columns names
df.columns
```

```python
# Feature Selection
x=df[['carbon_emissions', 'energy_output', 'renewability_index',
       'cost_efficiency']]
y=df['sustainability']
```

```python
# Split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# Initialize and train the model
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
# Prediction
y_pred=model.predict(x_test)
y_pred
```

```python
# Metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy=accuracy_score(y_test,y_pred)
print(f'The accuracy is: {accuracy}')
```

```python
# Confusion metrics
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True, cmap='coolwarm',xticklabels=['Not sustainable','Sustainable'],yticklabels=['Not sustainable','Sustainable'])
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion Metrics')
plt.show()
```

```python
cr=classification_report(y_test,y_pred,target_names=['Not sustainable','Sustainable'])
print(cr)
```

```python
# Feature importance
co=pd.DataFrame(model.coef_.T,index=x.columns,columns=['Coefficient'])
print(co)
```

```python
# Save the model
import joblib
joblib.dump(model,'LR_model.pkl')
```

```python
# Lets import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the dataset
df=pd.read_csv('agricultural sustainability.csv')
df.head()
```

```python
# Find the missing values
df.isnull().sum()
```

```python
# No. of rows and columns
df.shape
```

```python
# columns names
df.columns
```

```python
# Feature selection
x=df[['soil_health', 'crop_yield', 'water_usage', 'carbon_footprint',
       'fertilizer_use']]
y=df['sustainability']
```

```python
print(x.shape)
print(y.shape)
```

```python
# Split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Initialize and train the model
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
# Prediction
y_pred=model.predict(x_test)
print('The prediction are',y_pred)
```

```python
# Metrics
accuracy=accuracy_score(y_test,y_pred)
print(f'The accuracy is: {accuracy}')
```

```python
# Confusion metrics
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
# Classification report
cr=classification_report(y_test,y_pred,target_names=['Not sustainable','Sustainable'])
print(cr)
```

```python
# Feature importance
fi=pd.DataFrame(model.feature_importances_,index=x.columns,columns=['Importance'])
print(fi)
```

```python
# Save the model
import joblib
joblib.dump(model,'RFC_model.pkl')
```
