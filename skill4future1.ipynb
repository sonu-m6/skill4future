```python
class Car:
  wheels=4

  def __init__(self, make, year):
    self.make=make
    self.year=year

  def start_engine(self):
    print('Engine started!')

  @classmethod
  def get_wheels(cls):
    return cls.wheels

  @staticmethod
  def is_motor_vehicle():
    return True
```

```python
car = Car("Toyota",2022)
print(car.make)
print(car.year)
```

```python
car.start_engine()
```

```python
Car.get_wheels()
```

```python
Car.is_motor_vehicle()
```

```python
#Inheritance

class Vehicle:
  def __init__(self,brand,model):
    self.brand=brand
    self.model=model

  def display_info(self):
    return f"{self.brand} {self.model}"

class Car(Vehicle):
  def __init__(self,brand,model,color):
    super().__init__(brand,model)
    self.color=color

  def display_info(self):
    return f"{self.brand} {self.model} {self.color}"
```

```python
my_car=Car("Toyota","Camry","Red")
```

```python
my_car.display_info()
```

```python
import pandas as pd
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
df=pd.DataFrame(data)
df.head()
```

```python
null=df.isnull().sum()
null
```

```python
# Handling missing values
#1. Drop methon= dropna
#2. Impute method= fillna() mean,median,mode
```

```python
# drop method
handle=df.dropna()
handle
```

```python
df['Energy Consumption (MWh)'] = df['Energy Consumption (MWh)'].fillna(df['Energy Consumption (MWh)'].mean())
df['Cost (Million $)']=df['Cost (Million $)'].fillna(df['Cost (Million $)'].median())
df
```

```python
df['Cost per MWh'] = (df['Energy Consumption (MWh)'] / df['Cost (Million $)'])
display(df)
```

```python
#data preprocessing

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Energy Consumption (MWh)','Cost (Million $)']] = scaler.fit_transform(df[['Energy Consumption (MWh)','Cost (Million $)']])
display(df)
```

```python
#Standardization

from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
df[['Energy Consumption (MWh)','Cost (Million $)']] = std_scaler.fit_transform(df[['Energy Consumption (MWh)','Cost (Million $)']])
display(df)
```

```python
#OHE
encode_df=pd.get_dummies(df,columns=['Energy Source'],dtype=int)
encode_df
```

```python
#VISUALIZATION
#Line plot

months=['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption=[1200,1300,1100,1500,1400,1600]
```

```python
#lets import matplotlib

import matplotlib.pyplot as plt
plt.plot(months,energy_consumption, marker='o',linestyle='dashed',color='blue')
plt.xlabel('Months')
plt.ylabel('Energy Consumption (MWh)')
plt.title('Energy Consumption Over Months')
plt.show()
```

```python
#bar plot

energy_source=['Solar','Wind','Hydro','Geothermal','Biomass']
energy_consumption=[1200,1300,1100,1500,1400]
plt.bar(energy_source,energy_consumption,color='purple',edgecolor='black',linewidth=2)
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (MWh)')
plt.title('Energy Consumption by Source')
plt.show()
```

```python
#piechart

plt.pie(energy_consumption,labels=energy_source,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Energy Consumption by Source')
plt.show()
```

```python
#scatter plot

import matplotlib.pyplot as plt
energy_consumption=[1200,1300,1100,1500,1400,1600]
carbon_emission=[200,500,450,300,250,150]
plt.scatter(energy_consumption,carbon_emission,color='Red')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emission (kg)')
plt.title('Energy Consumption vs Carbon Emission')
plt.show()
```

```python
#customization

energy_source=['Solar','Wind','Hydro','Geothermal','Biomass']
energy_consumption=[1200,1300,1100,1500,1400]
plt.bar(energy_source,energy_consumption,color='yellow',edgecolor='black',linewidth=2)
plt.xlabel('Energy Source', fontsize=12)
plt.ylabel('Energy Consumption (MWh)',fontsize=12)
plt.title('Energy Consumption by Source')
plt.grid(True,linestyle='--',color='gray',alpha=0.5)
plt.legend(['Energy Consumption'],loc='upper left')
plt.show()
```
